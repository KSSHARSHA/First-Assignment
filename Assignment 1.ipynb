```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. Plotting X and Y

```python
x= np.arange(0,10)
y = x**2
plt.plot(x,y ,'r*', linewidth = 2 , markersize = 10)
plt.title('Assignment')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

# 2. Plotting the sales of two different groups in a week.

```python
days = [1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180]
sales_2 = [70,90,160,150,140,145,175]
plt.figure(figsize = (6,5),dpi =150)
plt.subplot(2,1,1)
plt.plot(days,sales_1,'go--')
plt.title('Sales of 1')
plt.xlabel('Days')
plt.ylabel('sales_1')
plt.figure(figsize = (6,5),dpi =150)
plt.subplot(2,1,2)
plt.plot(days,sales_2,'bo--')
plt.title('Sales of 2')
plt.xlabel('Days')
plt.ylabel('sales_2')
```
